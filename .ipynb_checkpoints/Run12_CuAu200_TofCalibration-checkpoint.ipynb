```python
import rootpy.ROOT as ROOT

# file containg final pass for QA
f = ROOT.TFile( "/Users/danielbrandenburg/bnl/local/data/Run12/cuau/t0_1/allhists.root", "READ" )
c = ROOT.TCanvas( "c", "c", 1000, 500 )
ROOT.gStyle.SetOptStat( 0 )
```

```python
c.Clear()
c.Divide( 2, 1 )
c.cd(1)
ROOT.gStyle.SetOptStat( 111 )
hVpd = f.zvertexVPD
hTpc = f.zvertex
hVpd.SetLineWidth( 2 )
hVpd.SetTitle( "VPD (Black), TPC (Red)" )
hVpd.Draw()
hTpc.SetLineColor( 'red' )
hTpc.SetLineWidth( 2 )
hTpc.Draw( "same" )

c.cd(2)
# Draw the difference between VPD and TPC
hDelta = f.zvertexDelta
hDelta.GetXaxis().SetRangeUser( -5, 5 )
fit = hDelta.Fit( "gaus", "S", "", -1, 1 )
hDelta.Draw()
hDelta.GetXaxis().SetTitle( "Z_{TPC} - Z_{VPD} [cm]" )
ROOT.gStyle.SetOptStat( 0 )
ROOT.gStyle.SetOptFit( 111 )
c
```
