```python
def plot( nn, f, name="eff_6_20_" ) :
    fitFunc = ROOT.TF1( "effFitFunc", "[0] * exp( - pow( [1] / x, [2] ) )", 0.0, 5.0 );
    fitFunc.SetParameters( .25, 0.05, 5.0 );
    fitFunc.SetLineWidth( 2 )
    fitFunc.SetLineColor( ROOT.kRed )
    g1 = f.Get( name )
    g1.GetXaxis().SetRangeUser( 0, 4 )
    g1.Fit( fitFunc, "R" )
    return g1.Clone( nn )
```

```python
c = Canvas( width=600, height=400 )
c.Divide( 3, 2 )

ROOT.gStyle.SetOptFit( 111  )
ROOT.gStyle.SetStatY( 0.8 );
ROOT.gStyle.SetStatX( 0.5 );
```

```python
c.Clear()
hp = plot( "per", f )
hp.Draw()
c.Print( "TOF_Eff_per.pdf" )
c
```
