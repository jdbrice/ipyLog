```python
import ROOT
# ROOT.enableJSVis()
```

```python
f1 = ROOT.TFile( "qm_nominal/PostCorr_P.root" )
f2 = ROOT.TFile( "BBC_MB_PostCorr_P_fd_left.root")
```

```python
def compare( plc, ch, cen ) :
    name = plc + "_yield/yield_" + plc + "_" + cen + "_" + ch
    
    h1 = f1.Get( name )
    h2 = f2.Get( name )
    
    cc = ROOT.TCanvas( "ccan_" + name, "ccan", 600, 800 )
    
    h1.Draw()
    h2.SetLineColor( ROOT.kRed )
    h1.Draw()
    h2.Draw("same")
    cc.SetLogy(1)
    cc.Draw()
    
    cc2 = ROOT.TCanvas( "cc2an_" + name, "ccan", 1000, 800 )
    hr = h1.Clone( "ratio_" + name )
    hr.Divide( h2 )
    hr.Draw()
    hr.Fit( "pol0", "WR", "", 0.6, 2.5 )
    hr.GetYaxis().SetRangeUser( 0.9, 1.1 )
    cc2.Draw()
    

    
    
    
```

## Comparing Spectra
1) QM Spectra used track-by-track corrections
2) BBC_MB spectra is using bin-center corrections
