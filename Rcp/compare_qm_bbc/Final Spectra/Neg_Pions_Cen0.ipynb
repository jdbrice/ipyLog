```python
TFile * f1 = new TFile( "PostCorr_Pi.root" );
TFile * f2 = new TFile( "BBC_MB_PostCorr_Pi.root" );
TCanvas * can = new TCanvas( "can", "can" );
```

```python
TH1D * pi_1 = (TH1D*)f1->Get( "Pi_yield/yield_Pi_0_n" );
TH1D * pi_2 = (TH1D*)f2->Get( "Pi_yield/yield_Pi_0_n" );
```

```python
pi_1->Draw();
pi_2->SetLineColor( kRed );
pi_2->Draw("same");
can->Draw();
can->SetLogy(1);
```

```python
TH1D * pi_ratio = (TH1D*)pi_1->Clone( "pi_ratio" );
pi_ratio->Divide( pi_2 );
```

```python



pi_ratio->Draw();
pi_ratio->Fit( "pol0" );
pi_ratio->GetYaxis()->SetRangeUser(0.8, 1.2);
can->Draw();
can->SetLogy(0);
```
