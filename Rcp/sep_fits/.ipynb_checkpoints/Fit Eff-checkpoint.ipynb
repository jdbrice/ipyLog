```python
%matplotlib inline
import rootpy.ROOT as R
R.gROOT.SetBatch(True)
from rootpy.io import root_open, DoesNotExist
from rootpy.plotting import Canvas, Hist, Legend
import rootpy.plotting.root2matplotlib as rplt
import matplotlib.pyplot as plt
```

```python

# if f_zd :
#     f_zd.Close()
# if f_zb :
#     f_zb.Close()

f_zd = root_open( "/Users/danielbrandenburg/bnl/local/work/rcpMaker/products/zd/Fit_tiny_K.root" )
f_zb = root_open( "/Users/danielbrandenburg/bnl/local/work/rcpMaker/products/zb/Fit_tiny_K.root" )
```

```python
def plot_yield( plc, pad ) :
    can.cd( pad )
    zb = f_zb.Get( plc + "_yield/yield_" + plc + "_0_p" ).Clone( "zb_" + plc )
    zb.SetMarkerColor( R.kBlue )
    zb.SetLineColor( R.kBlue )
    zb.Sumw2()
    zd = f_zd.Get( plc + "_yield/yield_" + plc + "_0_p" ).Clone( "zd_" + plc )
    zd.SetMarkerColor( R.kRed )
    zd.Sumw2()
    
    
    
    if "P" == plc :
        zd.GetYaxis().SetRangeUser( .1, 9  )
    if "Pi" == plc :
        zd.GetYaxis().SetRangeUser( zd.GetMaximum() * 1e-3, zd.GetMaximum() * .5  )
    
    zd.GetXaxis().SetRangeUser( 0, 2 )
    zd.SetMarkerStyle( 7 )
    zd.Draw()
    zb.Draw( "same" )
    zb.SetMarkerStyle( 7 )
    R.gPad.SetLogy(1)

def plot_eff( pad )
    can.cd( pad )
    can.SetLogy(0)
    eff = zb.Clone("eff")
    R.gStyle.SetOptStat(0)
    eff.Divide( zd )
    eff.GetYaxis().SetRangeUser( 0.8, 1.1 )
    eff.Draw()
```

```python
can = Canvas( width=1200, height= 900 )
can.Divide( 3, 2 )
plot_yield( "Pi", 1 )
plot_yield( "K", 2 )
plot_yield( "P", 3 )


can
```
