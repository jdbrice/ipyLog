```python
%matplotlib inline
import rootpy.ROOT as R
R.gROOT.SetBatch(True)
import numpy as np
from rootpy.io import root_open, DoesNotExist
from rootpy.plotting import Canvas, Hist, Legend
```

```python
def labelsForGEANT( hist ) :
    x = hist.GetXaxis()
    x.SetBinLabel( 1, "No ID" )
    x.SetBinLabel( 2, "e^{+}" )
    x.SetBinLabel( 3, "e^{-}" )
    x.SetBinLabel( 4, "#nu" )
    x.SetBinLabel( 5, "#mu^{+}" )
    x.SetBinLabel( 6, "#mu^{-}" )
    x.SetBinLabel( 7, "#pi" )
    x.SetBinLabel( 8, "#pi^{+}" )
    x.SetBinLabel( 9, "#pi^{-}" )
    x.SetBinLabel( 10, "K^{0}_{L}" )
    x.SetBinLabel( 11, "K^{+}" )
    x.SetBinLabel( 12, "K^{-}" )
    x.SetBinLabel( 13, "N" )
    x.SetBinLabel( 14, "P" )
    x.SetBinLabel( 15, "#bar{P}" )
    x.SetBinLabel( 16, "K^{0}_{S}" )
    x.SetBinLabel( 17, "#eta" )
    x.SetBinLabel( 18, "#Lambda" )
    x.SetBinLabel( 19, "#Sigma^{+}" )
    x.SetBinLabel( 20, "#Sigma^{0}" )
    x.SetBinLabel( 21, "#Sigma^{-}" )
    
```
