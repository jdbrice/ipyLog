```python
from IPython.html import widgets

def single_gauss( a, b, x ) :
    return a* np.exp( -b*b * x );

def plot_single_gauss( a, b ) :
    x = np.linspace( 0, 5 );
    plt.plot( x, single_gauss( a, b, x ) )
    plt.ylim( [0, 5] )
    
widgets.interact( plot_single_gauss, a=(0, 5), b=(0, 5, 0.001) )
```

```python
def double_gauss( a, b, c, d, x ) :
    return  a * np.exp( -b * x  ) + c * np.exp( -d*d * x );

def plot_double_gauss( a, b, c, d ) :
    x = np.linspace( 0, 5 );
    plt.plot( x, double_gauss( a, b, c, d, x ), x, single_gauss( a, b,  x ), x, single_gauss( c, d,  x ) )
    plt.ylim( [1e-3, 1] )
    plt.yscale('log')
    
widgets.interact( plot_double_gauss, a=(0.01, .5, 0.01), b=(0, 5, 0.001), c=(0.01, .5, 0.001), d=(0, 5, 0.001) )
```

```python
def supp_gauss( a, b, c, d, x ) :
    return  (1.0 - a * np.exp( -b * (x - 1)  )) * c * np.exp( -d*d * x );

def plot_supp_gauss( a, b, c, d ) :
    x = np.linspace( 0, 5 );
    plt.plot( x, supp_gauss( a, b, c, d, x ) )
    #plt.ylim( [1e-3, 1] )
    plt.yscale('log')
    
widgets.interact( plot_supp_gauss, a=(0.01, .5, 0.01), b=(0, 50, 0.1), c=(0.01, .5, 0.001), d=(0, 5, 0.001) )
```
