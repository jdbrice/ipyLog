```python
import rootpy.ROOT as ROOT

# file containg final pass
f = ROOT.TFile( "/Users/danielbrandenburg/bnl/local/data/Run12/cuau/t0_1/allhists.root", "READ" )
c = ROOT.TCanvas( "c", "c", 900, 500 )
```

```python
ROOT.gStyle.SetOptStat( 0 )

hVpd = f.zvertexVPD
hTpc = f.zvertex
hVpd.SetLineWidth( 2 )
hVpd.Draw()
hTpc.SetLineColor( 'red' )
hTpc.SetLineWidth( 2 )
hTpc.Draw( "same" )
c
```

```python
hDelta = f.zvertexDelta
hDelta.GetXaxis().SetRangeUser( -5, 5 )
fit = hDelta.Fit( "gaus", "S", "", -1, 1 )
hDelta.Draw()
c
```

```python
print "mean = ", fit.Value( 1 )
print "sigma = ", fit.Value( 2 )
```

* The mean should be less than ~10 mm from being centered at 0  
* sigma varies depending on energy and species
    * Run12 CuAu sigma = 0.72
