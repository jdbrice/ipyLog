##Starting Variables
The major variables that can be altered for this program. The sizes of the atoms gives the number of nucleons that will be placed within the atom. Impact_parameter is the vertical distance between the centers of the atoms. The nucleonxs is the radius of the nucleons. The tinc is the time increment that will be used for every update of the system. The x_locations are the locations of the two atoms.

```python
size1 =20
size2 =20
velocity1 = 200
velocity2 = 350
#impact_parameter = 30
nucleonxs = 30
lengthofatom1 = 40 
lengthofatom2 = 40
tinc = .001
skin_depth = 3
# Locations of x-coordinates for particle 1 and 2
radius = cross_to_radius(cross_section = nucleonxs)
x_location1 = -3
x_location2 = 3
num_iterations = 50
```

Comparison of the wood-saxon random generator to the graph of the wood saxon

```python
%time
ws =Wood_saxon(82,7,0,120)
array = []
for i in range(5000):
    num = ws.rand_woodsaxon()
    temp_var = np.random.uniform(-1,1)
    if (temp_var > 0):
        array.append(num*-1)
    else:
        array.append(num)
ws.complete_wood_saxon(res = 2000)
bw = 2
plt.hist(array,bins = (82*2 / bw), normed=True);
```
