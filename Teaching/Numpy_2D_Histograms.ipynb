## 2D Histograms

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 # Create a 2D Gaussian
n = 100000
x = np.random.randn(n)
y = np.random.randn(n)
 
# Estimate the 2D histogram
nbins = 100
H, xedges, yedges = np.histogram2d(x,y,bins=nbins, normed=True)
 
# Mask zeros
Hmasked = np.ma.masked_where(H<.05,H) # Mask pixels with a value of zero
 
# Plot 2D histogram using pcolor
fig1 = plt.figure()
plt.pcolormesh(xedges,yedges,H)
plt.xlabel('x')
plt.ylabel('y')
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')

# Draw the smae thing but mask out the bins with less than threshold
fig2 = plt.figure()
plt.pcolormesh(xedges,yedges,Hmasked)
plt.xlabel('x')
plt.ylabel('y')
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')

# Using imshow
fig3 = plt.figure()
plt.imshow(Hmasked)
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')
```
